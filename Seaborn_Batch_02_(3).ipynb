# ✨Data Pre-Processing





```python
#import the required libery's
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load seaborn
import seaborn as sns
```

# ✨ Scatter Plot

### Input Variable (x) --> total _bill column
### Output Variable (y) --> tip column

```python
#Load the pre-defined tips dataset of catering and staffing service based industry
#tips = sns.load_dataset("tips")
# Sample data

tips = pd.read_csv("tips.csv")
# Create a basic scatter plot
sns.scatterplot(x="total_bill",y="tip", data=tips)
plt.show()

```

# ✨ Custmize Plot in Seaborn

```python
# Costmize or update the scatterplot by adding context, style.
#set_style & set_context --> only for notebook
sns.scatterplot(x="total_bill",y="tip", data=tips)
sns.set_style("whitegrid")
sns.set_context("notebook")
plt.show()
```

# ✨Barplot & Count Plot

```python
# Plot the bar plot and with indivisual count for seperate visual (X,Y,data) as a argument,colomn name day & total_bill
sns.barplot(x="day", y="total_bill",data=tips)
plt.show()
#count has only one argument
sns.countplot(x="day",data=tips)
plt.show()
```

# ✨Histogram

```python
#plot the histogram which contents maximun 10 bins and known deviation as True
sns.histplot(x="total_bill", bins=10, kde=True , data=tips )
plt.show()
```

# ✨Pair Plot

```python
#plot the diffrent types of categories in single plot as easy to be visual. for coloum "sex" -->Gender
#hue is built in function is used
sns.pairplot(tips, hue="sex")
plt.show()
```

# ✨ Corelational Maxtix / Heatmap

```python
#Plot the corelational matrix
sns.heatmap(tips.corr(), annot= True, cmap="coolwarm")
plt.show()
```

# ✨Violin Plot

```python
sns.violinplot(x="day",y="total_bill",data=tips)
# Customize the plot (optional)


# Display the plot
plt.show()
```

# ✨ Kernal Density Estimate Plot

```python
sns.kdeplot(tips, fill=True)

# Add labels and title
plt.xlabel("X-axis Label")
plt.ylabel("Density")
plt.title("Kernel Density Plot")

plt.show()






```

# ✨ Joint Plot

```python
# Create a joint plot
sns.jointplot(x="total_bill", y="tip", data= tips, kind="scatter")
# Show the plot
plt.show()
```

# ✨ Rug Plot

```python
# Create a Rug Plot for the "total_bill" column
sns.rugplot(x="total_bill",data=tips, height= 0.5, color="blue")
# Set the style (optional)
plt.figure(figsize=(8,4))
# Set the figure size (optional)


# Add labels and a title (optional)


# Show the plot
plt.show()
```

# ✨Strip Plot

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.show()
```

# ✨ Box Plot

# ✨Boxen Plot

```python
sns.boxenplot(x="day",y="total_bill",data=tips)
# Add labels and a title


# Show the plot
plt.show()
```

# ✨ Regression Plot

```python
# Create a regression plot using lmplot
sns.lmplot(x="total_bill",y="tip",data=tips)

# Show the plot
plt.show()
```
